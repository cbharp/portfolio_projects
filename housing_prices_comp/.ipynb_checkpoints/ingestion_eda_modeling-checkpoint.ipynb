# Data Cleaning

### Mappings

```python
# take values from data dictionary and put them into a format that can be used to map in values when working with train and test
# data.
def read_data_description(file):
    with open(file) as f:
        outer_dict = {}
        counter = 0
        for line in f: 
            if counter == 0 and ':' in line and "story" not in line: # isolate lines from data dictionary that are columns ie MSSubClass
                field = line.split(":")[0]
                inner_dict = {}
                counter += 1
            if ':' in line and "story" not in line and counter>0: # column names contain a ":"
                outer_dict[field] = inner_dict
                field = line.split(":")[0]
                inner_dict = {}
            if ':' not in line or 'story' in line: # column values do not contain a ":"
                mapping = line.split('\t')
                if len(mapping)>1:
                    inner_dict[mapping[0].strip()] = mapping[1].strip('\n') 
            if field == 'SaleCondition' and len(inner_dict) == 6:
                outer_dict[field] = inner_dict
    return outer_dict
        
        
```

```python
outer_dict = read_data_description('data_description.txt')
```

```python
outer_dict
```

```python
# keeping all columns that are missing less than half of the total number of rows
def remove_missing_data_cols(df):
    columns_to_keep = df.isnull().sum().to_frame().rename(columns = {0:'null_value_count'})\
    .sort_values(by = 'null_value_count' ,ascending = False)
    half_of_total_rows = int(1460/2)
    columns_to_keep = columns_to_keep[columns_to_keep['null_value_count']<half_of_total_rows]
    columns_to_keep = columns_to_keep.reset_index()['index'].to_list()
    df = df[columns_to_keep]
    return df
```

```python
train_df = remove_missing_data_cols(train_df)
```

```python
train_df.shape
```

```python
# remove same columns from test df
test_df = test_df[train_df.loc[:,train_df.columns != 'SalePrice'].columns.to_list()]
```

```python
test_df.shape
```

```python
# see null values per column from most nulls to least 
train_df.isnull().sum().to_frame().rename(columns = {0:'null_value_count'})\
.sort_values(by = 'null_value_count' ,ascending = False)
```

```python
test_df.isnull().sum().to_frame().rename(columns = {0:'null_value_count'})\
.sort_values(by = 'null_value_count' ,ascending = False)
```

four columns were removed

```python
train_df = train_df.rename(columns = {'BedroomAbvGr':'Bedroom', # bedroom and kitchen in data description named differently
                                      'KitchenAbvGr':'Kitchen'})# than in training data so renaming to match
```

```python
# get list of columns that have values that are abbreviations 
map_columns = []
for column in train_df.columns:
    if column != 'SalePrice':
        for key in outer_dict[column].keys():
            if key:
                map_columns.append(column)
                break
```

```python
# cast list to set so search is quicker when seeing which fields are categorical or ordinal
map_columns = set(map_columns)
```

```python
# cast int columns that are categorical to string(nan for categorical field become string nan 
# but fine because they are cast back to float NaN when mapping in values). 
for column in train_df.columns:
    if column in map_columns:
        train_df[column] = train_df[column].apply(lambda x: str(x))
        train_df[column] = train_df[column].map(outer_dict[column]) # map in verbose version of data values for 
                                                                    # categorical fields
```

```python
train_df
```

```python
# impute missing continuous values with median
for column in train_df.columns:
    if column not in map_columns:
        train_df[column] = train_df[column].fillna(train_df[column].median())
```

```python
# impute missing categorical values with mode 
for column in list(map_columns):
    print(column)
    train_df[column] = train_df[column].fillna(train_df[column].mode()[0])
```

```python
train_df.isnull().sum().to_frame().rename(columns = {0:'null_value_count'})\
.sort_values(by = 'null_value_count' ,ascending = False)
```

### Dummy coding categorical variables

```python
train_df = pd.get_dummies(data = train_df, columns = list(map_columns))
```

```python
train_df
```

# Feature Selection and Engineering

```python
# separate dataset into indepent variable and target variable

x_data = train_df.drop('SalePrice', axis = 1)
y_data = train_df['SalePrice']
```

```python
# Use Random Forest Regressor to reduce number of features

model = RandomForestRegressor()
model.fit(x_data, y_data)
```

```python
importances = model.feature_importances_
```

```python
feature_importances_df = pd.DataFrame({
    'feature':x_data.columns,
    'importance':importances}).sort_values(by = 'importance', ascending = False)
```

```python
feature_importances_df
```

```python
# keeping features with at least %1 importance
threshold = .01
selected_features = feature_importances_df[feature_importances_df['importance']>threshold]
```

```python
selected_features
```

# Fit and interpret model 

```python
# subset x data for selected data
subset_x_train = x_data[selected_features['feature'].to_list()]
y_train = y_data
```

```python
model = LinearRegression()
model.fit(subset_x_train,y_train)
```
